c) Usuń stop-words z artykułu używając standardowej listy dla słów angielskich.
Podaj liczbę słów po tym etapie.

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

word_tokens = word_tokenize(text)

filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

filtered_sentence = []

for w in word_tokens:
	if w not in stop_words:
		filtered_sentence.append(w)

print(word_tokens)
print(filtered_sentence)
```

```python
print(len(filtered_sentence))
```

d) Sprawdź czy w naszym zestawie słów (bag of words) są jeszcze jakieś pominięte

```python
stop_words.add('On')
stop_words.add(',')
stop_words.add('In')
stop_words.add('We')
stop_words.add('At')
stop_words.add('.')
stop_words.add('The')
stop_words.add('``')
stop_words.add("'s")
stop_words.add("''")
```

e) Dokonaj lematyzacji dokumentu. Jaki lematyzer został wybrany? 

Alternatywnie:
możesz dokonać stemmingu. Podaj liczbę słów po tym etapie.

```python
from nltk.stem.wordnet import WordNetLemmatizer
lem = WordNetLemmatizer()
lemmatized_word = []
for i in range(0,len(filtered_sentence)):
  lemmatized_word.append(lem.lemmatize(filtered_sentence[i],"v"))
```

```python
print(lemmatized_word)
```

f) Podaj przetworzony dokument w formie wektora zliczającego słowa. 

Następnie
wyświetl na wykresie słupkowym 10 najczęściej występujących słów (oś X:
słowa, oś Y: liczba wystąpień słowa w tekście).

```python
from collections import Counter
counts = Counter(lemmatized_word)
print(counts)
```

```python
import matplotlib.pyplot as plt 
```

```python
slowa = []
liczba = []
for key,value in counts.items():
  slowa.append(key)
  liczba.append(value)
    
```

```python
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
```

```python
plt.bar(slowa[:10], liczba[:10])
plt.xlabel('Słowa')
plt.ylabel('Liczba wystąpień')
plt.rcParams["figure.figsize"] = (20,10)
plt.show()
```
