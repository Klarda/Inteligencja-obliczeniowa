```python
import matplotlib.pyplot as plt 
```

```python
ypoints = miasta['Gdansk']
xpoints = miasta['Rok']

plt.plot(xpoints,ypoints, marker = 'o', color='red')
plt.xlabel('Lata')
plt.ylabel('Liczba ludności [w tys.]')
plt.title('Ludność miasta Gdańsk')
plt.show()
```

d)Stwórz dodatkowo wykres, który będzie zestawiał zmiany ludności wszystkich miasta na 
jednym wykresie w różnych kolorach. Dodaj legendę.


```python
fig, ax = plt.subplots(figsize=(6, 6))

ax.set_title('Liczba ludnosci')
ax.plot(miasta['Rok'], miasta['Gdansk'], label='Gdansk', marker = 'o', color='red')
ax.plot(miasta['Rok'], miasta['Poznan'], label='Poznan', marker = 'o', color='green')
ax.plot(miasta['Rok'], miasta['Szczecin'], label='Szczecin',marker = 'o', color='blue')
plt.xlabel('Lata')
plt.ylabel('Liczba ludności [w tys.]')
ax.legend()
```
